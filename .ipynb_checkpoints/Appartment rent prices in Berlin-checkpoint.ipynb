```python
url = "https://www.immobilienscout24.de/Suche/S-T/P-{}/Wohnung-Miete/Berlin/Berlin"
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
```

```python
i = 1
results = list()
while True:
    r = requests.get(url.format(str(i)), headers={'User-Agent': user_agent})
    soup = BeautifulSoup(r.text, 'lxml')
    links = [tag.attrs['href'] for tag in soup.findAll('a', {'href' : re.compile("expose/[\d+]")})]
    results.extend(links)
    time.sleep(3)
    print(str(i))
    i += 1
```

```python
res = [res for res in results if "#" not in res]
res = list(set(res))
pd.Series(res).to_excel("immo_links.xlsx")
```

```python
results = []
import random
```

```python
for link in links[0]:
    print(link)
    html = requests.get('https://www.immobilienscout24.de' + link, headers={'User-Agent': user_agent})
    soup = BeautifulSoup(html.text, 'lxml')

    plz = soup.find("span", {'class' : 'zip-region-and-country'}).getText()
    headline = soup.find("h1", {'class' :  'font-semibold font-xl margin-bottom margin-top-m palm-font-l'}).getText()
    kalt = soup.find("dd", {'class' : 'is24qa-kaltmiete grid-item three-fifths' }).getText().strip()
    rooms = soup.find("dd", {'class' : 'is24qa-zimmer grid-item three-fifths' }).getText().strip()
    size = soup.find("dd", {'class' : 'is24qa-wohnflaeche-ca grid-item three-fifths' }).getText()
    warm = soup.find("dd", {'class' : 'is24qa-gesamtmiete grid-item three-fifths font-bold' }).getText()
    try:
        attr = soup.find("div", {'class' : 'criteriagroup boolean-listing padding-top-l'}).findAll("span")
    except:
        attr = None
    floor = soup.find("dd", {'class' : 'is24qa-etage grid-item three-fifths' })
    baujahr = soup.find("dd", {'class' : 'is24qa-baujahr grid-item three-fifths'})
    standard = soup.find("dd", {'class' : 'is24qa-qualitaet-der-ausstattung grid-item three-fifths'})
    results.append([plz, headline, kalt, rooms, size, warm, attr, baujahr, standard, link])
    pd.DataFrame(results).to_excel("immodata.xlsx")
    time.sleep(random.randint(3, 12))
```
